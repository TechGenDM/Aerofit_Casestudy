```python
# ── 3.1 Age Distribution ─────────────────────────────
fig, axes = plt.subplots(1, 2, figsize=(14, 4.5))
ax1 = axes[0]
ax1.hist(df["Age"], bins=18, color=C1, alpha=0.75, edgecolor="white", linewidth=1.3)
ax2t = ax1.twinx()
df["Age"].plot.kde(ax=ax2t, color=GOLD, linewidth=2.3)
ax2t.set_ylabel("Density", color=GOLD, fontsize=9); ax2t.tick_params(axis="y", colors=GOLD, labelsize=8)
ax2t.grid(False); ax2t.set_ylim(bottom=0); ax2t.spines["top"].set_visible(False)
ax1.axvline(df["Age"].mean(), color="#e74c3c", linestyle="--", linewidth=1.8, label=f"Mean: {df['Age'].mean():.1f}")
ax1.axvline(df["Age"].median(), color="#2ecc71", linestyle=":", linewidth=1.8, label=f"Median: {df['Age'].median():.0f}")
ax1.set_title("Customer Age Distribution — Overall"); ax1.set_xlabel("Age (years)"); ax1.set_ylabel("Count")
ax1.legend(fontsize=9); ax1.yaxis.grid(True); ax1.set_axisbelow(True)

ax3 = axes[1]
data = [df[df["Product"]==p]["Age"].values for p in ["KP281","KP481","KP781"]]
bp = ax3.boxplot(data, labels=["KP281\n(Entry)","KP481\n(Mid)","KP781\n(Premium)"],
                 patch_artist=True, medianprops=dict(color="white",linewidth=2.5),
                 whiskerprops=dict(linewidth=1.1,color=SLATE), capprops=dict(linewidth=1.1,color=SLATE),
                 flierprops=dict(marker="o",markersize=3.5,alpha=0.35))
for patch,color in zip(bp["boxes"],PAL): patch.set_facecolor(color); patch.set_alpha(0.7)
for i,(d,c) in enumerate(zip(data,PAL),1): ax3.scatter([i]*len(d),d,color=c,alpha=0.18,s=14,zorder=3)
ax3.set_title("Age Distribution by Product Tier"); ax3.set_xlabel("Product"); ax3.set_ylabel("Age (years)")
ax3.yaxis.grid(True); ax3.set_axisbelow(True)
plt.tight_layout(); plt.show()
```

```python
# ── 4.2 Conditional Probabilities ────────────────────
ct_row = pd.crosstab(df["Gender"], df["Product"], normalize="index")
ct_col = pd.crosstab(df["Gender"], df["Product"], normalize="columns")
print("  P(Product | Gender) — given a gender, probability of each product:")
display(ct_row.style.background_gradient(cmap="Blues",axis=1).format("{:.3f}")
    .set_table_styles([{"selector":"th","props":[("background","#1a2332"),("color","#c9943a"),("font-family","monospace"),("font-size","11px"),("padding","8px 14px")]},
                       {"selector":"td","props":[("font-family","monospace"),("font-size","11px"),("padding","7px 14px")]}]))
print("\n  P(Gender | Product) — given a product, probability of each gender:")
display(ct_col.style.background_gradient(cmap="Oranges",axis=0).format("{:.3f}")
    .set_table_styles([{"selector":"th","props":[("background","#1a2332"),("color","#c9943a"),("font-family","monospace"),("font-size","11px"),("padding","8px 14px")]},
                       {"selector":"td","props":[("font-family","monospace"),("font-size","11px"),("padding","7px 14px")]}]))

fig, ax = plt.subplots(figsize=(11, 4))
x=np.arange(3); w=0.32
b1=ax.bar(x-w/2,ct_row.loc["Female"]*100,width=w,color=C1,alpha=0.85,label="Female",edgecolor="white",linewidth=1.3)
b2=ax.bar(x+w/2,ct_row.loc["Male"]*100,width=w,color=C2,alpha=0.85,label="Male",edgecolor="white",linewidth=1.3)
for bs in [b1,b2]:
    for bar in bs:
        h=bar.get_height()
        ax.text(bar.get_x()+bar.get_width()/2,h+0.8,f"{h:.1f}%",ha="center",va="bottom",fontsize=9,fontweight="bold",color=NAVY)
ax.set_xticks(x); ax.set_xticklabels(["KP281 (Entry)","KP481 (Mid)","KP781 (Premium)"])
ax.set_title("P(Product | Gender) — Purchase Probability by Gender")
ax.set_ylabel("Probability (%)"); ax.set_ylim(0,75); ax.legend(); ax.yaxis.grid(True); ax.set_axisbelow(True)
plt.tight_layout(); plt.show()
```
